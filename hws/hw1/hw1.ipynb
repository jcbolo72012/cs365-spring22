```python
def strat_1(correct):
    test = 0000
    tries = 0
    while test != correct:
        test = np.random.randint(0,9999)
        tries +=1
    return tries

def strat_2(correct):
    tried = []
    test = 0000
    tries = 0
    while test != correct:
        while test in tried:
            test = np.random.randint(0,9999)
        tries +=1
        tried.append(test)
    return tries
```

```python
means1 = []
stds1 = []
means2 = []
stds2 = []
ts = [100,200,300,1000]
for t in ts:
    strat_1_100x = [strat_1(2022) for x in range(0,t)]
    print(strat_1_100x)
    s_1_100_avg = sum(strat_1_100x)/t
    s_1_100_std = np.std(np.asarray(strat_1_100x))

    means1.append(s_1_100_avg)
    stds1.append(s_1_100_std)

    print('strat_1:', s_1_100_avg , s_1_100_std)

    strat_2_100x = [strat_2(2022) for x in range(0,t)]
    s_2_100_avg = sum(strat_2_100x)/t
    s_2_100_std = np.std(np.asarray(strat_2_100x))
    means2.append(s_2_100_avg)
    stds2.append(s_2_100_std)
    print('strat_2:', s_2_100_avg , s_2_100_std)

```

```python
plt.figure()
plt.plot(ts,means1)
plt.errorbar(ts, means1, yerr = stds1)
plt.show()
plt.figure()
plt.plot(ts,means2)
plt.errorbar(ts, means2, yerr = stds2)
plt.show()
```
